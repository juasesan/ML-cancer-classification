```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
import itertools
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report
import ast
```

```python
df = pd.read_csv("cancer.csv")
```

```python
x = df.iloc[:,2:-1] # Features del modelo
y = df["diagnosis"] # Diagnóstico
```

### Separación de Datos

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
Xnorm=scaler.fit_transform(x_train)
```

```python
pca = PCA(n_components=10)
pca.fit(Xnorm)
```

```python
Xpca=pca.transform(Xnorm)
```

### Creación del modelo

```python
clf = MLPClassifier(random_state=0, max_iter=300)
```

### Tuning del Modelo

```python
param_grid = {
    'activation': ["identity","logistic","tanh","relu"],
    'hidden_layer_sizes': [(50,),(100,),(150,)],
    'max_iter': [50,100,200],
    'learning_rate_init': [0.01,0.001,0.0001]
}
```

```python
CV_rfc = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5)
```

```python
CV_rfc.fit(Xpca, y_train)
```

```python
CV_rfc.best_params_
```

```python
mlc = MLPClassifier(random_state=0, max_iter=200, hidden_layer_sizes=(50,), learning_rate_init=0.001,activation='relu')
```

```python
mlc.fit(Xpca, y_train)
```

### Evaluación del modelo

```python
Xnormtest = scaler.transform(x_test)
```

```python
Xpcatest = pca.transform(Xnormtest)
```

```python
mlc.score(Xpcatest,y_test)
```

```python
y_pred = mlc.predict(Xpcatest)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
labels = ['Benigno','Maligno']
fig,ax = plt.subplots(figsize=(7,5))
_=sns.heatmap(cm, annot=True, fmt='d', cmap="Reds", xticklabels=labels, yticklabels=labels, ax=ax)
_=ax.set_xlabel("Predicción", fontsize = 15)
_=ax.set_ylabel("Real", fontsize = 15)
```

```python
print(classification_report(y_test, y_pred))
```
